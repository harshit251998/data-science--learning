```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')  # seaborn consists of some in-built dataset 
```

```python
tips.head()
```

```python
 # ploting univariate data using distribution plot i.e. histogram
# to ignore the warnings 
from warnings import filterwarnings
sns.set_style('whitegrid')
sns.displot(tips['total_bill'],kde=False,bins=50,color='red') # bins is used to plot each instance of total_bills based on no. provided
```

```python
# joint plot used to show relationship b/w two variables
# from below scatter plot higher the bill higher the tip which shows tip is directly proportional to total_bill
sns.set_style('whitegrid')
sns.jointplot(x='total_bill',y='tip',data=tips,color='purple')
```

```python
# same above plot to give more attarctive look using kind in joint plot eg: hex (more the points darker the color of hexagone ),
#kde(kernel density estimation),reg
sns.set_style('whitegrid')
sns.jointplot(x='total_bill',y='tip',data=tips,color='orange',kind='hex')

#from below plot maximum tip is given b/w 10 and 20 dollor bill
```

```python
sns.set_style('whitegrid')
sns.jointplot(x='total_bill',y='tip',data=tips,color='orange',kind='reg')
#gives linear regression which is machine learning topic will learn later.
```

```python
# using pair plot which will plot all the numerical value of dataset and help to easily visualize the dataset just by passing dataset to it

sns.pairplot(tips,hue='sex',palette='coolwarm') # hue is used to plot non-numerical column in plot which is catagorical column
```
