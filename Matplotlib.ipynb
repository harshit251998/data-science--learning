```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
# functional use in plotting
plt.plot(x,y,color='purple',lw=1.5,linestyle='--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('square plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
# Object oriented plotting 

fig = plt.figure()  # it is like imaginary blank canvas , now we need to add axes to this canvas i.e. "fig"

axes = fig.add_axes([0,1,1,1]) # argument to be passed in axes (left,bottom,width,height of axes)

axes.plot(x,y,ls='-',lw=1,color='blue',marker='o',markersize=10,
         markerfacecolor='yellow')         # supported values for ls '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Square plot')
```

```python
fig=plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.4,0.5])
axes2 = fig.add_axes([0.6,0.1,0.4,0.5])

axes1.plot(x,y,ls='dashdot',lw=2,color='purple')
axes1.set_xlabel('x-axis')
axes1.set_ylabel('y-axis')
axes1.set_title('x^2 plot')

axes2.plot(y,x,ls='dashdot',lw=2,color='orange')
axes2.set_xlabel('x-axis')
axes2.set_ylabel('y-axis')
axes2.set_title('1/x^2 plot')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)  # easining  work of axes1 and axes2 as done in above cell
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,3))
axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()
```
